```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
df = pd.read_csv("airbnb_Oakland.csv")

# Filter hosts with latitude less than -110
filtered_df = df[df['Longitude'] < -110]

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Scatter plot with size based on revenue and color based on superhost status
scatter = ax.scatter(
    filtered_df['Longitude'], 
    filtered_df['Latitude'], 
    s=filtered_df['revenue'] / 100,  # Adjust the scaling factor for size
    c=['green' if superhost == 1 else 'red' for superhost in filtered_df['Superhost']],
    alpha=0.6,
    edgecolors='w',
    linewidth=1,
)

# Set axis labels and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Map of Hosts with Latitude Less Than -110\nColored by Superhost, Sized by Revenue')

# Create a legend
legend_labels = {'Superhost': 'green', 'Non-Superhost': 'red'}
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=label)
                   for label, color in legend_labels.items()]

ax.legend(handles=legend_elements, loc='upper right', title='Legend')

# Show the plot
plt.show()
```
