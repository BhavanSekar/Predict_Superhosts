```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import pandas as pd
df = pd.read_csv("airbnb_Oakland.csv")
# Assuming 'df' is your DataFrame
model = sm.OLS.from_formula('revenue ~ Superhost', data=df).fit()

plt.figure(figsize=(8, 6))
sns.lmplot(x='Superhost', y='revenue', data=df)
plt.title('Regression Plot of Revenue for Superhosts and Non-Superhosts')
plt.xlabel('Superhost')
plt.ylabel('Revenue')
plt.show()

# Print the model summary
print(model.summary())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Assuming 'df' is your DataFrame
model_price = sm.OLS.from_formula('available_days_aveListedPrice ~ Superhost', data=df).fit()

plt.figure(figsize=(8, 6))
sns.lmplot(x='Superhost', y='available_days_aveListedPrice', data=df)
plt.title('Regression Plot of Price for Superhosts and Non-Superhosts')
plt.xlabel('Superhost')
plt.ylabel('Price')
plt.show()

# Print the model summary
print(model_price.summary())
```
